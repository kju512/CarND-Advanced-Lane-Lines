## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

## Cell One

```python

import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#%matplotlib qt
%matplotlib inline
# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(8,5,0)
objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

# Make a list of calibration images
images = glob.glob('./camera_cal/calibration*.jpg')

# Step through the list and search for chessboard corners
for fname in images:
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6),None)

    # If found, add object points, image points
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)

        # Draw and display the corners
        img = cv2.drawChessboardCorners(img, (9,6), corners, ret)
        plt.imshow(img)
```

## Cell Two

```python
ret,mtx,dist,rvecs,tvecs=cv2.calibrateCamera(objpoints,imgpoints,gray.shape[::-1],None,None)

#undistort image function
def imgUndist(img,mtx,dist):
    h,w=img.shape[:2]
    newcameramtx, roi=cv2.getOptimalNewCameraMatrix(mtx,dist,(w,h),1,(w,h))
    # undistort
    img_undist = cv2.undistort(img, mtx, dist, None,newcameramtx)

    # crop the image
    x,y,w,h = roi
    imgdst = img_undist[y:y+h, x:x+w]
    return imgdst

img= cv2.imread('./camera_cal/calibration2.jpg')
imgud= imgUndist(img,mtx,dist)
#cv2.imwrite('calibration1result.png',imgud)

plt.figure(num=0,dpi=600)
plt.subplot(1,2,1)
plt.title('Raw')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Undistortion')
plt.imshow(imgud)
```

## Cell four

```python
# create a mask the owner by color.
def GetBinaryByHLSColor(img,thresh=(70, 255)):
    img = np.copy(img)
    # Convert to HLS color space and separate the S channel
    hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS).astype(np.float)
    s_channel = hls[:,:,2]
    
    # Threshold color channel
    s_binary = np.zeros_like(s_channel)
    s_binary[(s_channel >= thresh[0]) & (s_channel <= thresh[1])] = 1
    
    return s_binary

imgbyhlscolor= GetBinaryByHLSColor(imgud,thresh=(220, 255))
plt.imshow(imgbyhlscolor,cmap='gray')
```

## Cell five

```python
# create a mask the owner by color.
def GetBinaryByHSVColor(img, thresh=(215, 255)):
    img = np.copy(img)
    # Convert to HLS color space and separate the S channel
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    v_channel = hsv[:,:,2]
    
    # Threshold color channel
    s_binary = np.zeros_like(v_channel)
    s_binary[(v_channel >= thresh[0]) & (v_channel <= thresh[1])] = 1
    
    return s_binary

imgbyhsvcolor= GetBinaryByHSVColor(imgud,thresh=(220, 255))
plt.imshow(imgbyhsvcolor,cmap='gray')
```

## Cell six

```python
# create a mask the owner by color.
def GetBinaryByLabColor(img, thresh=(208, 255)):
    img = np.copy(img)
    # Convert to YCrCb color space and separate the S channel
    Lab = cv2.cvtColor(img, cv2.COLOR_BGR2Lab)
    L_channel = Lab[:,:,0]
    
    # Threshold color channel
    s_binary = np.zeros_like(L_channel)
    s_binary[(L_channel >= thresh[0]) & (L_channel <= thresh[1])] = 1
    
    return s_binary

imgbyLabcolor= GetBinaryByLabColor(imgud,thresh=(208, 255))
plt.imshow(imgbyLabcolor,cmap='gray')
```

## Cell seven

```python
# create a mask the owner by Sobelx.
def GetBinaryBySobel(img, direct='x',thresh=(8,255)):
    hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
    s_channel = hls[:,:,2]
    # Sobel x
    if direct=='x':
        sobel = cv2.Sobel(s_channel, cv2.CV_64F, 1, 0) # Take the derivative in x
    else:
        sobel = cv2.Sobel(s_channel, cv2.CV_64F, 0, 1) # Take the derivative in y
    abs_sobel = np.absolute(sobel) # Absolute derivative to accentuate lines away from horizontal
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    
    # Threshold gradient
    sbinary = np.zeros_like(scaled_sobel)
    sbinary[(scaled_sobel >= thresh[0]) & (scaled_sobel <= thresh[1])] = 1
    
    return sbinary


imgbysobelx= GetBinaryBySobel(imgud,direct ='x',thresh=(30,255))
plt.imshow(imgbysobelx,cmap='gray')
```

```python
### 
def GetBinaryBySobelMag(img, sobel_kernel=7,thresh=(20,255)):
    
    # Apply the following steps to img
    # 1) Convert to grayscale
    hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
    s_channel = hls[:,:,2]    
    # 2) Take the gradient in x and y separately
    sobelx=cv2.Sobel(s_channel,cv2.CV_64F,1,0,ksize=sobel_kernel)
     # 3) Calculate the magnitude 
    sobely=cv2.Sobel(s_channel,cv2.CV_64F,0,1,ksize=sobel_kernel)
    # 4) Scale to 8-bit (0 - 255) and convert to type = np.uint8    
    sobel=np.sqrt(sobelx*sobelx+sobely*sobely)
    # 5) Create a binary mask where mag thresholds are met 
    sobelmask=np.uint8(255*sobel/np.max(sobel))
    binary_output=np.zeros_like(sobelmask)
    binary_output[(sobelmask>=thresh[0])&(sobelmask<=thresh[1])]=1
    # 6) Return this mask as your binary_output image
    return binary_output
imgbysobelmag= GetBinaryBySobelMag(imgud,sobel_kernel=21,thresh=(80,255))
plt.imshow(imgbysobelmag,cmap='gray')
```

## Cell ten

```python
def GetBinaryBySobeldir(img, sobel_kernel=7,thresh=(0, np.pi/2)):
    
    # Apply the following steps to img
    # 1) Convert to grayscale
    gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    # 2) Take the gradient in x and y separately
    sobelx=cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=sobel_kernel)
    sobely=cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=sobel_kernel)
    # 3) Take the absolute value of the x and y gradients
    sobelx_abs=np.absolute(sobelx)
    sobely_abs=np.absolute(sobely)
    # 4) Use np.arctan2(abs_sobely, abs_sobelx) to calculate the direction of the gradient  
    angle=np.arctan2(sobely_abs,sobelx_abs)
    # 5) Create a binary mask where direction thresholds are met
    binary_output=np.zeros_like(angle)
    binary_output[(angle>=thresh[0])&(angle<=thresh[1])]=1
    # 6) Return this mask as your binary_output image
    return binary_output
imgbysobeldir= GetBinaryBySobeldir(imgud,sobel_kernel=15,thresh=(0.2, 1.2))
plt.imshow(imgbysobeldir,cmap='gray')
```

```python
def GetPerspectiveTransformImg(img):
    img_size = (img.shape[1], img.shape[0])
    src = np.float32(
                    [[(img_size[0] / 2) - 80, img_size[1] / 2 + 100],
                    [((img_size[0] / 6) - 10), img_size[1]],
                    [(img_size[0] * 5 / 6) + 10, img_size[1]],
                    [(img_size[0] / 2 + 80), img_size[1] / 2 + 100]])
    dst = np.float32(
                    [[(img_size[0] / 4), 0],
                    [(img_size[0] / 4), img_size[1]],
                    [(img_size[0] * 3 / 4), img_size[1]],
                    [(img_size[0] * 3 / 4), 0]])
    #Compute the perspective transform, M, given source and destination points
    M = cv2.getPerspectiveTransform(src, dst)
    Minv = cv2.getPerspectiveTransform(dst, src)
    #Warp the combinated binary image using the perspective transform, M
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_NEAREST)
    return warped,M,Minv
binary_warped,M,Minv=GetPerspectiveTransformImg(imgcomb)
plt.imshow(binary_warped,cmap='gray')
```

## Cell sixteen

```python
def FitLinebyWindowSliding(binary_warped, nwindows = 9, margin = 60, minpix = 5):
    # Assuming you have created a warped binary image called "binary_warped"
    # Take a histogram of the bottom half of the image
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[50:midpoint])
    rightx_base = np.argmax(histogram[midpoint:histogram.shape[0]-50]) + midpoint

    # Set height of windows
    window_height = np.int(binary_warped.shape[0]/nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated for each window
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        
        # Identify the nonzero pixels in x and y within the window
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xleft_low) & (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xright_low) & (nonzerox < win_xright_high)).nonzero()[0]
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        # If you found > minpix pixels, recenter next window on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

    # Concatenate the arrays of indices
    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)

    # Extract left and right line pixel positions
    left_x = nonzerox[left_lane_inds]
    left_y = nonzeroy[left_lane_inds] 
    right_x = nonzerox[right_lane_inds]
    right_y = nonzeroy[right_lane_inds] 

    # Fit a second order polynomial to each
    if left_x.size==0:
        left_fit = np.array([0,0,0], dtype='float')
    else:
        left_fit = np.polyfit(left_y, left_x, 2)
    if right_x.size==0:
        right_fit = np.array([0,0,0], dtype='float')
    else:
        right_fit = np.polyfit(right_y, right_x, 2)

    # Generate x and y values for plotting
    #ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    #left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    #right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    
    return left_fit,right_fit,left_x,left_y,right_x,right_y
```

## Cell eighteen

```python
# Define conversions in x and y from pixels space to meters
#ym_per_pix = 30/720 # meters per pixel in y dimension
#xm_per_pix = 3.7/700 # meters per pixel in x dimension
#y_eval is which location We want to Calculate Lane Curve Rad
def CalculateLaneCurveRadAndVehicleOffset(y_eval,fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700):
    # Generate x and y values for 
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    fitx = fit[0]*ploty**2 + fit[1]*ploty + fit[2]
    # Fit new polynomials to x,y in world space
    fit_cr = np.polyfit(ploty*ym_per_pix, fitx*xm_per_pix, 2)
    # Define y-value where we want radius of curvature
    # I'll choose the maximum y-value, corresponding to the bottom of the image
    # Calculate the new radii of curvature
    curverad = ((1 + (2*fit_cr[0]*y_eval*ym_per_pix + fit_cr[1])**2)**1.5) / np.absolute(2*fit_cr[0])   
    #calculate the vehicle line_base_pos
    line_base_pos=(fitx[-1]-binary_warped.shape[1]/2.0)*xm_per_pix
    return curverad,line_base_pos
# Now our radius of curvature is in meters
y_eval=binary_warped.shape[0]-1
left_curverad,offset_l=CalculateLaneCurveRadAndVehicleOffset(y_eval,left_fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700)
right_curverad,offset_r=CalculateLaneCurveRadAndVehicleOffset(y_eval,right_fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700)
print(left_curverad, 'm', right_curverad, 'm', (left_curverad+right_curverad)/2, 'm',(offset_l+offset_r)/2,'m')
# Example values: 632.1 m    626.2 m
```

## Cell twenty

```python
#Add the dectected lane region to the raw image
def AddDectectedLaneToImg(imgundist,binary_warped, left_fit, right_fit, Minv):
    # Create an image to draw the lines on
    warp_zero = np.zeros_like(binary_warped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

    # Generate x and y values 
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    
    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))

    #draw boundary of the lane
    cv2.polylines(color_warp,np.array(pts_left, np.int32),False,(0,0,255),20)
    cv2.polylines(color_warp,np.array(pts_right, np.int32),False,(0,0,255),20)
    
    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))

    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, Minv, (binary_warped.shape[1], binary_warped.shape[0])) 
    # Combine the detected lane with the original image
    output = cv2.addWeighted(imgundist, 1.0, newwarp, 0.8, 0)
    
    #Calculate the lane curvation radius and the offset of the vehicle
    y_eval=binary_warped.shape[0]-1
    left_curverad,offset_l=CalculateLaneCurveRadAndVehicleOffset(y_eval,
                                                 fit=left_fit,
                                                 ym_per_pix = 30/720,
                                                 xm_per_pix = 3.7/700)
    right_curverad,offset_r=CalculateLaneCurveRadAndVehicleOffset(y_eval,
                                                 fit=right_fit,
                                                 ym_per_pix = 30/720,
                                                 xm_per_pix = 3.7/700)
    center_curverad=(left_curverad+right_curverad)/2
    offset=(offset_l+offset_r)/2
    # put the text on the output image
    text="Radius of the curvature of the lane ={:.0f}m".format(center_curverad)
    loc=(output.shape[0]//2-150, 50) #the location where the text will be showed
    fontFace=cv2.FONT_HERSHEY_COMPLEX  #text font
    fontScale=1.1
    tcolor=(0,255,0)#text color
    thickness = 2
    lineType = 4
    cv2.putText(output,text,loc,fontFace,fontScale,tcolor,thickness)
    #cv2.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2)
    if offset < 0:
        text="Vehicle is {:.2f}m left of the lane center".format(np.absolute(offset))
        loc=(output.shape[0]//2-150, 80)
        cv2.putText(output,text,loc,fontFace,fontScale,tcolor,thickness)
    else:
        text="Vehicle is {:.2f}m right of the lane center".format(np.absolute(offset))
        loc=(output.shape[0]//2-150, 80)
        cv2.putText(output,text,loc,fontFace,fontScale,tcolor,thickness)
    return output
result=AddDectectedLaneToImg(imgud,binary_warped, left_fit, right_fit, Minv)
plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))
```

## Cell twenty one

```python
# Define a class to receive the characteristics of each line detection
class Line():
    def __init__(self):
        # was the line detected in the last iteration?
        self.detected = False  
        # x values of the last n fits of the line
        self.recent_xfitted = [] 
        #average x values of the fitted line over the last n iterations
        self.bestx = None     
        #polynomial coefficients averaged over the last n iterations
        self.best_fit = None  
        #polynomial coefficients for the most recent fit
        self.current_fit = []
        #radius of curvature of the line in some units
        self.radius_of_curvature = None 
        #distance in meters of vehicle center from the line
        self.line_base_pos = None 
        #difference in fit coefficients between last and new fits
        self.diffs = np.array([0,0,0], dtype='float') 
        #x values for detected line pixels
        self.allx = None  
        #y values for detected line pixels
        self.ally = None
        #threshold for update 
        self.thesh_c=np.array([1.0e-3,1.0,5.0e+3], dtype='float') 
        #iteration 
        self.iter_n=3
        
```

## Cell twenty two

```python
#use all aboved method to get a pipe line
#def pipeline(img,mtx,dist,leftLine,rightLine):
def pipeline(img):
    #undistort the image by the matrix and coefficent
    imgundist=imgUndist(img,mtx,dist)
    #get a binary image by saturation channel of hls color
    imgbyhlscolor=GetBinaryByHLSColor(imgundist, thresh=(200, 255))
    #get a binary image by vlaue channel of hsv color
    imgbyhsvcolor=GetBinaryByHSVColor(imgundist, thresh=(200, 255))
    #get a binary image by vlaue channel of hsv color
    imgbyLabcolor=GetBinaryByLabColor(imgundist, thresh=(208, 255))
    #get a binary image by gradient magnitude of Sobel mothed
    imgbysobelmag=GetBinaryBySobelMag(imgundist,sobel_kernel=21,thresh=(80,255))
    #get a binary image by gradient x of Sobel mothed
    imgbysobelx=GetBinaryBySobel(imgundist,direct ='x',thresh=(30,255))
    #get a binary image by gradient y of Sobel mothed
    imgbysobely=GetBinaryBySobel(imgundist,direct ='y',thresh=(30,255))
    #create a combination binary image from all of aboved binary image
    imgcomb=np.zeros_like(imgbysobelmag)
    imgcomb[(imgbyhlscolor==1)|(imgbyhsvcolor==1)|(imgbyLabcolor==1)|(imgbysobelmag==1)\
        |(imgbysobelx==1)|(imgbysobely==1)]=1
    #warpe the combined image by perspectiveTransforming
    binary_warped,M,Minv=GetPerspectiveTransformImg(imgcomb)
    
    left_fit,right_fit,left_x,left_y,right_x,right_y=FitLinebyWindowSliding(binary_warped)
    leftLine.detected = True
    rightLine.detected = True
    leftLine.allx=leftx
    leftLine.ally=lefty
    rightLine.allx=rightx
    rightLine.ally=righty
    
    #store the left current fit to leftLine object
    #determin whether the Line need to be updated  
     #consider the lastest iter_n 
    if len(leftLine.recent_xfitted)==0:
        leftLine.recent_xfitted.append(left_fitx)
        leftLine.bestx=left_fitx
        leftLine.current_fit.append(left_fit)
        leftLine.best_fit=left_fit
        
        y_eval=binary_warped.shape[0]-1 
        left_curverad,offset=CalculateLaneCurveRadAndVehicleOffset(y_eval,leftLine.best_fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700)
        leftLine.radius_of_curvature = left_curverad 
        leftLine.line_base_pos = offset
        #leftLine.diffs = 0
    elif len(leftLine.recent_xfitted)>=leftLine.iter_n:
        leftLine.diffs = leftLine.best_fit-left_fit
        if np.max(np.absolute(leftLine.diffs/leftLine.thesh_c))<0.5:
            leftLine.recent_xfitted.pop(0)
            leftLine.recent_xfitted.append(left_fitx)
            leftLine.bestx=np.mean(np.array(leftLine.recent_xfitted),axis=0)
        
            leftLine.current_fit.pop(0)
            leftLine.current_fit.append(left_fit)
            leftLine.best_fit=0.95*leftLine.best_fit+0.05*left_fit
        
            y_eval=binary_warped.shape[0]-1 
            left_curverad,offset=CalculateLaneCurveRadAndVehicleOffset(y_eval,leftLine.best_fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700)
            leftLine.radius_of_curvature = left_curverad 
            leftLine.line_base_pos = offset
        else:
            leftLine.current_fit.append(leftLine.best_fit)
            leftLine.detected = False
    else:  #len(leftLine.recent_xfitted)<5&>=2
        leftLine.diffs = leftLine.best_fit-left_fit
        if np.max(np.absolute(leftLine.diffs/leftLine.thesh_c))<0.5:
            leftLine.recent_xfitted.append(left_fitx) 
            leftLine.bestx=np.mean(np.array(leftLine.recent_xfitted),axis=0)
        
            leftLine.current_fit.append(left_fit)
            leftLine.best_fit=0.95*leftLine.best_fit+0.05*left_fit 
        
            y_eval=binary_warped.shape[0]-1 
            left_curverad,offset=CalculateLaneCurveRadAndVehicleOffset(y_eval,leftLine.best_fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700)
            leftLine.radius_of_curvature = left_curverad 
            leftLine.line_base_pos = offset
        else:
            leftLine.current_fit.append(leftLine.best_fit)
            leftLine.detected = False

            
    #store the right current fit to rightLine object
    if len(rightLine.recent_xfitted)==0:
        rightLine.recent_xfitted.append(right_fitx)
        rightLine.bestx=right_fitx
        rightLine.current_fit.append(right_fit)
        rightLine.best_fit=right_fit
        
        y_eval=binary_warped.shape[0]-1 
        right_curverad,offset=CalculateLaneCurveRadAndVehicleOffset(y_eval,rightLine.best_fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700)
        rightLine.radius_of_curvature = right_curverad 
        rightLine.line_base_pos = offset
        #rightLine.diffs = 0 

    elif len(rightLine.recent_xfitted)>=rightLine.iter_n:
        rightLine.diffs = rightLine.best_fit-right_fit
        if np.max(np.absolute(rightLine.diffs/rightLine.thesh_c))<0.5:
            rightLine.recent_xfitted.pop(0)
            rightLine.recent_xfitted.append(right_fitx)
            rightLine.bestx=np.mean(np.array(rightLine.recent_xfitted),axis=0)
        
            rightLine.current_fit.pop(0)
            rightLine.current_fit.append(right_fit)
            rightLine.best_fit=0.95*rightLine.best_fit+0.05*right_fit  
        
            y_eval=binary_warped.shape[0]-1 
            right_curverad,offset=CalculateLaneCurveRadAndVehicleOffset(y_eval,rightLine.best_fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700)
            rightLine.radius_of_curvature = right_curverad 
            rightLine.line_base_pos = offset
        else:
            rightLine.current_fit.append(rightLine.best_fit)
            rightLine.detected = False  
    else:  #len(rightLine.recent_xfitted)<5&>=2
        rightLine.diffs = rightLine.best_fit-right_fit
        if np.max(np.absolute(rightLine.diffs/rightLine.thesh_c))<0.5:
            rightLine.recent_xfitted.append(right_fitx) 
            rightLine.bestx=np.mean(np.array(rightLine.recent_xfitted),axis=0)
        
            rightLine.current_fit.append(right_fit)
            rightLine.best_fit=0.95*rightLine.best_fit+0.05*right_fit 
        
            y_eval=binary_warped.shape[0]-1 
            right_curverad,offset=CalculateLaneCurveRadAndVehicleOffset(y_eval,rightLine.best_fit,ym_per_pix = 30/720,xm_per_pix = 3.7/700)
            rightLine.radius_of_curvature = right_curverad 
            rightLine.line_base_pos = offset
        else:
            rightLine.current_fit.append(rightLine.best_fit)
            rightLine.detected = False
    outputimg=AddDectectedLaneToImg(imgundist,binary_warped,leftLine.best_fit,rightLine.best_fit, Minv)
    
    return outputimg
leftLine=Line()
rightLine=Line()
 
result=pipeline(img)
plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))
```
